#### Bullet 4: Experiment with word embeddings

```python
import hw4a as h
rnn_mod, idx2word = h.test_rnnslu_embeddings(nhidden=200, win=7, emb_dimension=50, normal=True, nepochs=60)
```

```python
def findBestWord(oute, idx2word, indx, indx2):
    # Adding two vectors
    x = oute.emb.get_value()[indx,]+oute.emb.get_value()[indx2,]
    # Computing their cosine similarity
    tmp = oute.emb.get_value().dot(x)
    # Have to remove the index itself when selecting, 
    #    otherwise it'll select itself as it's best possible pair
    smps = [ri for ri in range(572) if ri not in (indx,indx2)]
    val = tmp[smps].argmax()
    print("Most similar word for ("+idx2word[indx]+"+"+idx2word[indx2]+
           ') is '+idx2word[smps[val]]+ ' with similarity of '+ str(tmp[smps][val]))
    
def PairCompare(oute, idx2word, indx1, indx2, indx3, indx4, plus=True):
    # Comparing the cosine similarity when adding two word-vectors
    if plus:
        x1 = oute.emb.get_value()[indx1,]+oute.emb.get_value()[indx2,]
        x2 = oute.emb.get_value()[indx3,]+oute.emb.get_value()[indx4,]
        outv = x1.dot(x2)  
        print("("+idx2word[indx1]+" + "+idx2word[indx2]+') is ('+ 
               idx2word[indx3] + ' + '+idx2word[indx4] +') with similarity of '+str(outv))
    # Comparing the cosine similarity when subtracting two word-vectors
    else:
        x1 = oute.emb.get_value()[indx1,]-oute.emb.get_value()[indx2,]
        x2 = oute.emb.get_value()[indx3,]-oute.emb.get_value()[indx4,]
        outv = x1.dot(x2)  
        print("("+idx2word[indx1]+" - "+idx2word[indx2]+') is ('+ 
               idx2word[indx3] + ' - '+idx2word[indx4] +') with similarity of '+str(outv))
```

```python
# I chose these because my name is Francisco and my brother's name is Jose (and I thought these were interesting)
PairCompare(rnn_mod, idx2word, 205, 147, 247, 415, plus=True)  # All related tocalifornia and high cosine similarity
PairCompare(rnn_mod, idx2word, 205, 247, 275, 93, plus=True)   # Not very similar while the pairs are similar
PairCompare(rnn_mod, idx2word, 205, 247, 77, 100, plus=True)   # Very similar since both paris are cities
PairCompare(rnn_mod, idx2word, 137, 113, 128, 477, plus=False) # Somewhat similar though I'm surprised this isn't higher
PairCompare(rnn_mod, idx2word, 26, 36, 80, 115, plus=False)    # This pairing is meaningless
PairCompare(rnn_mod, idx2word, 156, 570, 542, 80, plus=True)   # This pairing is meaningless
```

```python
# Interesting results from finding the most similar word for two pairs of words summed together
findBestWord(rnn_mod, idx2word, 275, 245) # Pretty interesting
findBestWord(rnn_mod, idx2word, 275, 472) # Pretty interesting
findBestWord(rnn_mod, idx2word, 205, 247) # Makes sense, popular cities
findBestWord(rnn_mod, idx2word, 224, 176) # Also makes sense
findBestWord(rnn_mod, idx2word, 31, 134)  # Less interesting but not unreasonable
findBestWord(rnn_mod, idx2word, 26, 80)   # Less interesting but not unreasonable
```

#### Discussion

My RNN was able to learn the parity function for both 8 and 12 bits (as well as for 2,3,4, and 5--this can be verified by reviewing my previous commits). 

Finding a model to learn this function actually took quite a bit of effort, which made the entire exercise rather interesting. Fortunately, due to the time extension, I was able to figure out what was going on. 

My learning rate was set too high before and so I was simply not taking large enough steps in the direction of the gradient, so I would timeout at various local optimums (or at least what my model thought were optimums). In theory, increasing the number of epochs could have potentially gotten the model to converge but this would have certainly taken too long. Also, other heuristics/exceptions may have interfered with a model converging with too small of a learning rate(e.g., the other stopping criteria in the **test\_rnn\_parity** function). I think this was useful because it emphasized a point taught previously in lecture: *choosing the appropriate learning rate matters!* 


I tried various hyperparameter combinations to tune the model appropriately. This included implementing an RNN with embedding (**test\_rnn\_parity**) and without embedding (**test\_rnn\_parity2**), I ultimately removed the embedding option and deleted it from my homework code, but my former **test\_rnn\_parity** function exists in previous commits. My original thought was that longer training time could have helped learn the function, but, as stated above, I realize now that this would be insufficient given some of the stopping rules in the code.
