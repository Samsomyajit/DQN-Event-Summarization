```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def plotdata():
    sd = pd.read_csv('./sim_perf.txt', sep=';')
    sd['loss'] = sd['loss'].astype(float)
    sd['epsilon'] = sd['epsilon'].astype(float)
    plt.figure(figsize=(16,8))
    plt.title("Rouge and Loss as a function of time")
    plt.plot(sd['epoch'], sd['loss'], c='red')
    plt.plot(sd['epoch'], sd['rouge'], c='blue')
    plt.plot(sd['epoch'], sd['epsilon'], c='gray')
    plt.ylim([0,1])
    plt.grid()
    plt.show()
    return sd
```
